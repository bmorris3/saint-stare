```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
```

```python
table = Table.read('data/target_list_saintex_sel.csv', delimiter=';')
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
ax.hist(table['Spt'])
xticks = [8, 9, 10, 11, 12, 13]
xticklabels = ['M8', 'M9', 'L0', 'L1', 'L2', 'L3']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)
ax.set_xlabel('Spectral Type')
ax.set_ylabel('Counts')
for s in ['right', 'top']: 
    ax.spines[s].set_visible(False)
fig.savefig('plots/sptypes.pdf', bbox_inches='tight')
```

```python
plt.figure(figsize=(4, 3))
ax = plt.subplot(111, projection='hammer')

ax.scatter(np.radians(table['ra_deg']) - np.pi, np.radians(table['dec_deg']), marker='.')

ax.grid(ls=':', color='gray')

ax.set_xticks(np.radians(np.arange(-120, 150, 60)))

# for s in ['right', 'top']: 
#     ax.spines[s].set_visible(False)
fig.savefig('plots/coords.pdf', bbox_inches='tight')
```
